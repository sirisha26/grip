 From the given ‘Iris’ dataset, predict the optimum number of clusters and represent it visually.

Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df.head()
#Load Iris dataset and see the 1st 5 rows
```

```python
iris_df.describe()
```

Defining target and predictor

```python
x=iris.data[:, :2]
y=iris.target
```

Visualizing data using scatter plots

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.xlabel('Sepal Lenght',fontsize=16)
plt.ylabel('Sepal Width',fontsize=16)
```

```python
# Data Visualization
sns.pairplot(iris_df)
```

# Finding the optimum number of clusters for k-means classification

```python
x = iris_df.iloc[:, [0, 1, 2, 3]].values

from sklearn.cluster import KMeans
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

# Applying kmeans to the dataset

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++',
                max_iter = 400, n_init = 40, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

# Visualising the clusters 

```python
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], 
            s = 60, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], 
            s = 60, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1],
            s = 60, c = 'green', label = 'Iris-virginica')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            s = 60, c = 'yellow', label = 'Centroids')

plt.legend()
```

We can conclude that from above visualizations 3 is the optimum value of K in this iris dataset
