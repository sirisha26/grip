### In this regression task we will predict the percentage of marks that a student is expected to score based upon the number of hours they studied. This is a simple linear regression task as it involves just two variables.

```python
# import pandas library for reading csv file and matplotlib for plotting graphs
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read the data from the given link 
data = pd.read_csv('http://bit.ly/w-data')
print(data)
```

```python
# Plotting the data 
data.plot(x='Hours', y='Scores',style='o', label = "Score")
plt.xlabel('Hours studied')
plt.ylabel('Scores of students')
plt.title('Hours VS Scores')
plt.legend()
plt.show()
```

```python
# Separate the independent and dependent values from the data.
X = data[['Hours']]
y = data['Scores']
```

```python
# Divide the training set and test set from the given data.
train_X = X[:20]
train_y = y[:20]

test_X = X[20:]
test_y = y[20:]
```

```python
# Plotting training data
plt.scatter(train_X,train_y,color='r')
plt.xlabel('Hours studied')
plt.ylabel('Scores of students')
plt.title('Hours VS Scores of Training Set')
plt.show()
```

```python
# Import LinearRegression from sklearn.linear_model and train the training set.
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
regr.fit(train_X,train_y)
print('Training Completed Successfully!')
```

```python
# Plot the training set and the line we got from our training.
line = regr.intercept_ + regr.coef_*train_X 

plt.scatter(train_X, train_y)
plt.plot(train_X, line)
plt.show()
```

```python
# Predict the score of test set and display the predicted as well as actual scores.
predicted_y = regr.predict(test_X)
df = pd.DataFrame({'Predicted Score':predicted_y, 'Actual Score':test_y})
print(df)
```

```python
# Comparison between predicted and actual scores of test set.
plt.scatter(test_X,predicted_y,label = "Predicted",color='g')
plt.scatter(test_X,test_y,label = "Actual",color='r')
plt.title("Predicted and Actual Scores Comparison")
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
plt.legend()
plt.show()
```

```python
#Predict score for 9.25 hours of study
hours = 9.25
pred_hours = regr.predict([[hours]])
print("No of Hours Studied by student = {}".format(hours))
print("Predicted Score = {}".format(pred_hours[0]))
```

```python
# Mean Absolute Error and the r-square value
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(test_y, predicted_y)) 
print('R squared value:',metrics.r2_score(test_y,predicted_y))
```
